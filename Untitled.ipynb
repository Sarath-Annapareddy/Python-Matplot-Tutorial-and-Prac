```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
x1=np.linspace(0,5,10)
y1=x1**2
plt.plot(x1,y1)
plt.title("First plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(["cdacaas"],loc=0)
```

Printing Multiple plots

```python
plt.subplot(2,2,1)
plt.tight_layout(pad=3.5)
plt.plot(x1,y1,"ro--")
plt.xlabel("sacc")
plt.subplot(2,2,2)
plt.plot(x1,y1,"bo--")
plt.ylabel("ncjasnc")
plt.subplot(2,2,3)
plt.plot(x1,y1,"go--")
plt.ylabel("ncjasnc")
plt.subplot(2,2,4)
plt.plot(x1,y1,"mo--")
plt.ylabel("ncjasnc")
```

```python
fig1,axes =plt.subplots(figsize=(10,4),nrows=2,ncols=3)

plt.tight_layout()
plt.tight_layout()
plt.tight_layout()
axes[0,2].set_title("plot1")
axes[0,2].set_xlabel("x-axis")
axes[0,2].set_ylabel("y-axis")
axes[0,2].plot(x1,y1,label="nothing")
axes[0,2].set_xticks([0,2,4,6])


axes[0,2].legend()
```

```python
fig2=plt.figure(figsize=(7,4))
ax2=fig2.add_axes([0.1,0.1,0.9,0.9])
ax2.set_title("plot3")
ax2.set_xlabel("X-Axis")
ax2.set_ylabel("Y-Axis")
ax2.plot(x1,y1,"r",label="My Plot")
ax2.legend(loc=5)

ax3=fig2.add_axes([0.2,0.5,0.45,0.4])
ax3.set_title("plot4")
ax3.set_xlabel("x-Axis")
ax3.set_ylabel("y-Axis")
ax3.plot(x1,y1,label="My Plot")
ax3.legend(loc=0)

```
