### Functional Plot (plt.plot())

```python
### using arrays or list makes no differenece

x=[0,1,2,3,4,5]
y=[0,1,4,9,16,25]
x_arr=np.array(x)
y_arr=np.power(x,2)

plt.plot(x,y)
plt.plot(x_arr,y_arr)
```

### Using Figure Objects
figure is an object that contains all the plot elements, and axes. When we create a figure we get decide its size in inches

```python
fig1=plt.figure(figsize=(6,4))             # created figure of size (5*4)
ax=fig1.add_axes((0.1,0.1,0.9,0.9))       # added an axesto the figure (left,bottom,width,height)
ax.set_xlabel("X-Axis")                   # set_xlabel just like plt.xlabel
ax.set_ylabel("Y-Axis")
ax.set_title("Plot")
ax.plot(x_arr,y_arr,color="r",label="red")
ax.plot(y_arr,x_arr,color="orange",label="orange")

### creating another axes in the same figure
ax1=fig1.add_axes((0.5,0.5,0.3,0.3))
ax1.plot(x_arr,y_arr,color="black",label="black")

### setting legend
ax1.legend(loc=0)
ax.legend(loc=0)

ax.text(10,22,"Message")                 # keeping a message in the plot
```

Printing Multiple Plots using SUBPLOTS 

```python
fig2,ax2= plt.subplots(figsize=(6,5),nrows=2,ncols=3)    
plt.tight_layout()                                     # creating gap between plots to avoid ovarlap of plots  

### Accessing sublot position using indexing
ax2[0,0].plot(x_arr,y_arr,color="red")
ax2[1,1].plot(y_arr,x_arr,color="blue")
ax2[0,2].plot(x_arr,y_arr,color="green")
```

### Appearance Options

```python
# Default colors (b: blue, g: green, r: red, c: cyan, m: magenta,
# y: yellow, k: black, w: white)
# color="0.75" creates a 75% gray
# You can use hexcodes color="#eeefff"
# You can use color names found next like this color="burlywood"
# https://en.wikipedia.org/wiki/Web_colors
# alpha defines the percentage of opacity

# The default line width is 1, so to double it put in 2 and so forth

# There are many line styles 
# matplotlib.org/3.1.0/gallery/lines_bars_and_markers/linestyles.html
# You can also provide a sample like '-.'

# Markers can mark your provided points on the graph
# https://matplotlib.org/3.3.0/api/markers_api.html

#Add a grid, color, dashes(5pts 1 pt dashes separated by 2pt space)

fig3=plt.figure(figsize=(5,4))             
ax3=fig3.add_axes((1,1,1,1)) 
ax3.plot(x_arr,y_arr,marker="o",markerfacecolor="red",markersize="8")
ax3.grid(color="black",dashes=(5,2,1,2))    
ax3.set_facecolor("burlywood")
```

#### Barcharts - Example 2

```python
# Show percentages of males & females in engineering
m_eng = (76, 85, 86, 88, 93)
f_eng = (24, 15, 14, 12, 7)
branches=["CSE","IT","ECE","EEE","MECH"]

# Get evenly spaced values for each interval
spc = np.arange(0,len(m_eng))

## plotting men-percentages on the bar chart
fig5=plt.figure(figsize=(5,3))
ax5=fig5.add_axes([1,1,1,1])
ax5.bar(spc,m_eng,width=0.4,label="Male")

## plotting female-percentages on the bar chart next to male percentages
ax5.bar(spc+0.4,f_eng,width=0.4)

# Define x ticks in the middle of the 2 bars and define their names
#plt.xticks(spc+0.4/2,branches,rotation='horizontal',color="red",size=15)

ax5.set_xticks(spc+0.4/2)
ax5.set_xticklabels(branches,color="red",rotation=45)

## Annotate
for i,j,k in zip(spc,m_eng,f_eng):
       plt.annotate(str(j),xy=(i,j),xytext=(i,j+1))
       plt.annotate(str(k),xy=(i+0.4,k),xytext=(i+0.4,k+1))
```

### Heat Maps

```python
# A heatmap is a color coded representation of data from a 2D list
symptoms = ["Coronavirus","Influenza","Pneumonia","Dyspnea"]
dates = ["Jun28","Jul05","Jul12","Jul19","Jul26","Aug02","Aug09","Aug16","Aug21"]
symp_per = np.array([[5.2, 5.5, 5.7, 5.6, 5.3, 5.1, 5.0, 4.9, 5.3],
                    [3.5, 4.0, 4.3, 3.9, 3.5, 3.2, 2.7, 2.2, 2.0],
                    [1.8, 2.2, 2.3, 2.2, 2.1, 1.9, 1.7, 1.4, 1.3],
                    [1.0, 1.1, 1.1, 1.0, 0.9, 0.8, 0.8, 0.8, 0.7]])

fig_10, axes_10 = plt.subplots()
# Dfine data to use and color map
im = axes_10.imshow(symp_per, cmap="Wistia")

# Add ticks at data points and labels
axes_10.set_xticks(np.arange(len(dates)))
axes_10.set_yticks(np.arange(len(symptoms)))
axes_10.set_xticklabels(dates)
axes_10.set_yticklabels(symptoms)

# Rotate labels on the bottom so they don't overlap
plt.setp(axes_10.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(symptoms)):
    for j in range(len(dates)):
        text = axes_10.text(j, i, symp_per[i, j],
                       ha="center", va="center", color="k",fontweight="bold")
        
```
